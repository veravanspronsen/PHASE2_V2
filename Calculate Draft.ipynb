```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
mass = [60.75* 10 **3, 65.61* 10**3]  # kg

radius = 3.0/2
A = 3 * np.pi * radius**2 # m^2
rho = 1025 # kg/m^3

```

```python
for i in range(len(mass)):
    draft = mass[i] / (rho * A)
    print(f"Draft: {draft:.2f} m for mass {mass[i]/1000:.2f} tons")
```
