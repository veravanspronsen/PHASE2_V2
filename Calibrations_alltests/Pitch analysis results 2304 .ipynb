```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# =========================
# INPUT
# =========================
excel_path = Path(r"C:\Users\verav\Desktop\Studie\Afstuderen\Decay_calibration_outputs\calibrated_pitch_decay_coefficients.xlsx")
sheet_name = "pitch"

own_linear_damping = 80

# =========================
# DATA INLADEN
# =========================
df = pd.read_excel(excel_path, sheet_name=sheet_name)

construction_col = "construction"
lin_col = "linear_damping"
quad_col = "quadratic_damping"

# =========================
# PLOT 1: LINEAIR vs KWADRATISCH
# =========================
plt.figure(figsize=(8, 6))

marker_map = {
    "fixedwith": ("o", "green", "Fixed with Deck"),
    "fixedwithout": ("x", "blue", "Fixed without Deck"),
    "spring": ("*", "red", "Spring with Deck"),
}

# 👉 correcte manier: per groep plotten
for construction, (marker, color, label) in marker_map.items():
    sub = df[df[construction_col] == construction]

    plt.scatter(
        sub[lin_col],
        sub[quad_col],
        s=80,
        marker=marker,
        color=color,
        label=label
    )


plt.xlim([0, 350])
plt.ylim([-50, 1400])
plt.xlabel("Lineaire damping coefficient")
plt.ylabel("Kwadratische damping coefficient")
plt.title("Lineair vs Kwadratisch (gecalibreerd)")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
