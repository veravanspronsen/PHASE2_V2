```python
import pandas as pd
import matplotlib.pyplot as plt

TITLE_SIZE = 16
LABEL_SIZE = 14
TICK_SIZE = 12

files = [
    (r"C:\Users\verav\Desktop\Studie\Afstuderen\PHASE2_v2\PHASE2_V2\Heave\Fixedwith\Heave_fixedwithdeck.xlsx", "fixed with deck", "o"),
    (r"C:\Users\verav\Desktop\Studie\Afstuderen\PHASE2_v2\PHASE2_V2\Heave\Fixedwithout\Heave_fixedwithoutdeck_summary.xlsx", "     fixed without deck", "s"),
    (r"C:\Users\verav\Desktop\Studie\Afstuderen\PHASE2_v2\PHASE2_V2\Heave\Spring\Heave_spring_summary.xlsx", "spring", "o")
]
```

```python
# Create a combined DataFrame from all files
combined_df = pd.DataFrame()

for file, label, marker in files:
    df_temp = pd.read_excel(file)
    df_temp['group'] = label
    df_temp['marker'] = marker
    combined_df = pd.concat([combined_df, df_temp], ignore_index=True)

# Now plot using the combined DataFrame
plt.figure(figsize=(8, 6))

for group_label in combined_df['group'].unique():
    group_data = combined_df[combined_df['group'] == group_label]
    x = group_data["fine_lin"]
    y = group_data["fine_quad"]
    size = abs(group_data["A0_signed [m]"]) * 100  # scale amplitude to point size
    marker = group_data['marker'].iloc[0]  # assuming marker is the same for the group
    
    plt.scatter(x, y, s=size, marker=marker, alpha=0.6, label=group_label)

plt.xlabel("Linear damping coefficient", fontsize=LABEL_SIZE)
plt.ylabel("Quadratic damping coefficient", fontsize=LABEL_SIZE)
plt.title("Calibrated damping coefficients heave", fontsize=TITLE_SIZE)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
